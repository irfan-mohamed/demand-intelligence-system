## Data Understanding

The data contains in two tables.
- products.csv (informations on products)
- Sales.csv (informations on sales)

products.csv

| S. No | Feature | Description |
|-------|---------|------|
| 1 | product_id | Unique identifier of product |
| 2 | product_name | Name of the product |
| 3	| unit | Measurement unit |
| 4	| product_type |	Type of the product	|
| 5 | brand_name | Name of the products brand |
| 6	| manufacturer_name | Name of the manufacturer |
| 7 | l0_category | Name of main category of the product |
| 8	| l1_category | Name of the subcategory |
| 9	| l2_category | Name of the subcategory |
| 10 | l0_category_id | Id of main category |
| 11 | l1_category_id | Id of the subcategory |
| 12 | l2_category_id | Id of the subcategory |

Sales.csv

| S. No | Feature | Description |
|-------|---------|------|
| 1 | id | Unique identifier for each transaction row |
| 2 | unnamed_0_1 | Redundant index column (generated during data export, can be dropped) |
| 3	| unnamed_0 | Redundant index column (duplicate of index, can be dropped) |
| 4	| date_ |	Date when the order was placed |
| 5 | city_name | City where the order was delivered or placed |
| 6	| order_id | Unique identifier for each order (one order can have multiple products) |
| 7 | cart_id | Identifier for the shopping cart/session |
| 8	| dim_customer_key | Unique identifier for the customer |
| 9	| procured_quantity | Quantity of the product purchased in the order |
| 10 | unit_selling_price | Selling price per unit of the product |
| 11 | total_discount_amount | Total discount applied to the product |
| 12 | product_id | Unique identifier for the product |
| 13 | total_weighted_landing_price | Total cost incurred for the product (including procurement and logistics cost) |

## Final Insights.

There need some data cleanups.

------------------------------------------------------------
1. Column Cleanup
------------------------------------------------------------
- Drop redundant columns:
  - products.csv - Unnamed: 0
  - sales.csv - unnamed_0, unnamed_0_1
- Rename inconsistent columns:
  - date_ - order_date
- Standardize column naming

------------------------------------------------------------
2. Handling Missing Values
------------------------------------------------------------
products.csv:
- brand_name - fill with 'Unknown' or use most frequent brand
- manufacturer_name - fill with 'Unknown' or derive from brand if possible

sales.csv:
- total_weighted_landing_price → 79,355 NULLs
  - Option 1: Impute using product-level average cost
  - Option 2: Estimate using % of selling price
  - Option 3: Drop rows if impact is minimal

------------------------------------------------------------
3. Data Type Validation
------------------------------------------------------------
- Convert date_ - DATE format
- Ensure numeric columns:
  - procured_quantity - INTEGER
  - unit_selling_price - FLOAT
  - total_discount_amount - FLOAT
  - total_weighted_landing_price - FLOAT
- Ensure categorical columns are STRING type


------------------------------------------------------------
4. Data Consistency Checks
------------------------------------------------------------
- Ensure:
  - procured_quantity ≥ 0
  - unit_selling_price ≥ 0
  - total_discount_amount ≥ 0
- Validate:
  - selling_price ≥ discount


------------------------------------------------------------
5. Standardization & Formatting
------------------------------------------------------------
- Trim spaces in text fields:
  - product_name, brand_name, categories
- Convert text to consistent case (lower/upper)

------------------------------------------------------------
6. Referential Integrity Checks
------------------------------------------------------------
- Ensure:
  - sales.product_id exists in products.product_id
- Remove or flag orphan records
